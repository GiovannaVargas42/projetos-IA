```python
from google.colab import drive
import os

# 1. Montando o Google Drive
drive.mount('/content/drive')

# 2. Defina o caminho do diretório onde os arquivos PNG estão localizados
diretorio = '/content/drive/TrabalhoIA/Taurinos'  # Substitua com o caminho do seu diretório no Google Drive

# 3. Listando todos os arquivos PNG no diretório
arquivos_png = [f for f in os.listdir(diretorio) if f.endswith('.png')]

# 4. Exibindo a lista de arquivos PNG
if arquivos_png:
    print(f"Arquivos PNG encontrados: {arquivos_png}")

    for arquivo in arquivos_png:
      file_path = os.path.join(diretorio, arquivo)
      os.remove(file_path)  # Exclui o arquivo


else:
    print("Nenhum arquivo PNG encontrado no diretório.")
```
